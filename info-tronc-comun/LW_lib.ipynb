Tri selection

```python
def minimum(t,j=0):
    min = j
    for k in range(j+1,len(t)):
        if t[k]<t[min]: #le tri est stable
            min = k
    return min

def tri_selection(t):
    for j in range(len(t)-1):
        i = minimum(t,j) #le tri est en place
        if i!=j :
            t[i],t[j] = t[j], t[i]
```

```python
from random import randint

t = [randint(0,10**1) for _ in range(10**1)] #[8,6,1,4,2]
tri_selection(t)
print(t)
```

Graph function matplotlib.pyplot

```python
import numpy as np
import matplotlib.pyplot as plt

def graph_func(func,a=-10,b=10,n=100):
    x = np.linspace(a,b,n)
    plt.plot(x,func(x))

def show_graph(x="x",y="y(x)"):
    plt.xlabel(x)
    plt.ylabel(y)
    plt.grid(linestyle='-.')
    plt.show
```

```python
graph_func(np.sin,-np.pi,np.pi)
graph_func(np.cos,-np.pi,np.pi)
test = lambda x : x**2
graph_func(test,-np.pi,np.pi)
graph_func(lambda x : x+x/2 ,-np.pi,np.pi)

show_graph(y='test')
```

Excution time tester

```python
from time import time

def timer(func):
    def wraper(*args, **kwargs):
        t = time()
        val = func(*args, **kwargs)
        print(f"function : {func.__name__}    with args {args} and kwargs {kwargs}    took {time()-t}s to run")
        return val
    return wraper
```

```python
from time import sleep

@timer
def test(a,b=None):
    sleep(2)
    return a

print(test("val",b=0))
```
